So I divided the dataframe into target variable and independent variables

```python
inputs = df.drop('Star type', axis='columns') #this is independent variables
output = df['Star type'] # this is a target variable
```

To see the inputs dataframe

```python
inputs.head()
```

To see the output dataframe

```python
output.head()
```

Check the unique value of 'Star color' column and its frequency

```python
inputs['Star color'].value_counts()
```

Also check the unique value of 'Spectral Class' column and its frequency

```python
inputs['Spectral Class'].value_counts()
```

From those two labels above we see that in 'Star color' label has some redundant of color name, such as 'Blue-white' and 'Blue White' and so more. We should clear it first.

```python
inputs['Star color'].replace({'Blue White':'Blue-white',
                         'Blue white':'Blue-white',
                         'Blue-White':'Blue-white',
                         'yellowish':'Yellowish',
                         'white':'White',
                         'Blue ':'Blue',
                         'Blue white ':'Blue-white'}, inplace = True)
```

Check again the unique value of 'Star color' column. We see that it is clear.

```python
inputs['Star color'].value_counts()
```

Now by this point we all know that machine learning algorithms can only work on numbers, they cannot understand the labels. So what we have to do is to convert 'Star color' and 'Spectral Class' columns into numbers and one of the ways is to use Label Encoder from sklearn.preprocessing

```python
from sklearn.preprocessing import LabelEncoder
```

Then I'm going to create the object of those class (starcolor and spectralclass)

```python
le_starcolor = LabelEncoder()
le_spectralclass = LabelEncoder()
```

After I create those object now look at inputs dataframe that created before, I make extra column in those two dataframe using fit and transform method

```python
inputs['Star color new'] = le_starcolor.fit_transform(inputs['Star color'])
inputs['Spectral Class new'] = le_starcolor.fit_transform(inputs['Spectral Class'])
```

Once you have create that we see that the inputs dataframe has two extra columns, I have labels and coded those into numbers

```python
inputs.head()
```

The next step is to drop those label columns

```python
inputs_n = inputs.drop(['Star color','Spectral Class'], axis='columns')
inputs_n.head(6)
```

Label encoded transformation

```python
# Red = 5
# Blue = 0
# Blue-white = 1
# yellow-white = 12
# White = 6
# yellow-white = 11
# Yellowish = 9
# Yellowish White = 10
# Whitish = 8
# Orange = 2
# White-Yellow = 7
# Pale yellow orange = 4
# Orange-Red = 3

# M = 5
# B = 1
# O = 6
# A = 0
# F = 2
# K = 4
# G = 3
```

Now we are ready to train this classifier. For decision tree algorithms I import this from sklearn library

```python
from sklearn import tree
```

Create the model

```python
model = tree.DecisionTreeClassifier()
```

Perlu dilakukan train test split method..

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs_n,output, test_size = 0.2)
```

Train the model above

```python
X_test
```

```python
X_train
```

```python
model.fit(X_train,y_train)
```

The score of our model

```python
model.score(X_train,y_train)
```

Finally the model is ready, now I want to predict what start type a star which have following characteristics : Temperature = 43000 (K), Luminosity = 8.43200, Radius = 2.7200, Absolute magnitude = 9.380, Star color = Blue (encoded to 0) and Spectral Class = F (encoded to 2).

```python
model.predict([[43000,8.43200,2.7200,9.380,0,2]])
```
